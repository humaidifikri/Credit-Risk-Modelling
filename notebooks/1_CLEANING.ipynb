```python
import pandas as pd
```

```python
df = pd.read_csv('../data/loan_data_2007_2014.csv')
df.head()
```

```python
df.info()
```

```python
df.describe(include='all')
```

```python
df.duplicated().sum()
```

```python
for c in df.columns.to_list():
    print(f'{c}: ',df[c].isna().sum()/len(df)*100)
```

```python
for c in df.columns.to_list():
    if df[c].isna().sum()/len(df)*100 > 70.0:
        print(f'{c}: ',df[c].isna().sum()/len(df)*100)
```

```python
many_nan_col = []
for c in df.columns.to_list():
    if df[c].isna().sum()/len(df)*100 > 70.0:
        many_nan_col.append(c)
print(many_nan_col)
```

```python
df.drop(columns=many_nan_col,inplace=True)
df.info()
```

```python
profile_cols = ['Unnamed: 0','id','member_id','url','title','zip_code','addr_state','policy_code','pymnt_plan']
df.drop(columns=profile_cols,inplace=True)
df.info()
```

```python
leakage_cols = ['out_prncp', 'out_prncp_inv', 'total_pymnt', 'total_pymnt_inv',
    'total_rec_prncp', 'total_rec_int', 'total_rec_late_fee',
    'recoveries', 'collection_recovery_fee',
    'last_pymnt_d', 'last_pymnt_amnt']
df.drop(columns=leakage_cols,inplace=True)
df.info()
```

```python
df_cat = df.select_dtypes(include='str')
for c in df_cat.columns.to_list():
    print(f'{c}: ',df_cat[c].nunique())
```

```python
df_cat = df.select_dtypes(include='str')
for c in df_cat.columns.to_list():
    if df_cat[c].nunique() >= 10 or df_cat[c].nunique() < 2:
        df.drop(columns=c,inplace=True)
df.info()
```

```python
cat_cols = df.select_dtypes(include='str')
print([cat_cols[c].value_counts() for c in cat_cols])
```

```python
nan_cols = []
for c in df.columns.to_list():
    if df[c].isna().sum() > 0:
        nan_cols.append(c)
nan_cols
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(20,10))
fig = 1

for c in nan_cols:
    plt.subplot(3,5,fig)
    sns.kdeplot(data=df[nan_cols],x=c,)
#     plt.hist(x=df[c])
    plt.tight_layout
    plt.xlabel(c)
    fig+=1
```

```python
for c in nan_cols:
    before = df[c].skew()
    
    median_val = df[c].median()
    df[c] = df[c].fillna(median_val)
    
    after = df[c].skew()
    print(f"{c}: skew before={before:.3f} | after={after:.3f}")
```

```python
df['home_ownership'] = df['home_ownership'].replace(
    {'NONE': 'OTHER', 'ANY': 'OTHER'}
)

df['home_ownership'].value_counts()
```

```python
df.isna().sum()
```

```python
status_map = {
    # Good Loan
    'Fully Paid':                                               0,
    'Current':                                                  0,
    'Does not meet the credit policy. Status:Fully Paid':       0,

    # Bad Loan
    'In Grace Period':                                          1,
    'Late (16-30 days)':                                        1,
    'Late (31-120 days)':                                       1,
    'Default':                                                  1,
    'Charged Off':                                              1,
    'Does not meet the credit policy. Status:Charged Off':      1,
}

df['target'] = df['loan_status'].map(status_map)

df = df[df['target'].notnull()]
df.head()
```

```python
df.drop(columns='loan_status',inplace=True)
df.to_csv('../data/cleaned_loan_data',index=False)
```
