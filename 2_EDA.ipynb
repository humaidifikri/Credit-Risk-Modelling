```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import PowerTransformer
```

```python
df = pd.read_csv('../data/cleaned_loan_data')
df.head()
```

```python
sns.countplot(data=df,x='target');
```

```python
sns.countplot(data=df,x='grade',hue='target');
```

```python
fig, ax = plt.subplots(figsize=(8, 4))
for label, color, name in zip([0, 1], ['steelblue', 'tomato'], ['Good Loan', 'Bad Loan']):
    df[df['target'] == label]['loan_amnt'].plot.kde(ax=ax, label=name, color=color, linewidth=2)
# sns.histplot(data=df,x='loan_amnt',hue='target');
```
